Tg Graphs for 35 °C and 45 °C:

```python
env=[35,45]
cor=['bo--','go--']
incx=[50,50]
incy=[1,1]
dimen=[(4,5),(5,6)]

for index, value in enumerate(env):
    plt.figure(figsize=dimen[index])
    plt.xlabel('Exposure Time (h)')
    plt.ylabel('Tg (°C)')
    
    xint = range(0, math.ceil(max(df_dsc[f"Tempo ({value})"].dropna()))+50,incx[index])
    yint = range(math.floor(min(df_dsc[f"Tg ({value})"].dropna())), math.ceil(max(df_dsc[f"Tg ({value})"].dropna()))+1,incy[index])
    plt.xticks(xint)
    plt.yticks(yint)
    plt.plot(df_dsc[f"Tempo ({value})"].dropna().astype(int),df_dsc[f"Tg ({value})"].dropna(),cor[index])
    plt.savefig(f'G:\\Meu Drive\\TCC\\resultados ensaios\\DSC\\Grafs_TCC_Images\\Tg{value}.png', dpi = 500)
```

Crystallinity Degree for 35 °C, 45 °C, and T-H-R:

```python
env=[35,45,'TUR']
cor=['bo--','go--','ro--']
delta_max = [1,2,5]
delta_min = [0,0,2]

inc=[1,2,4]

dimen=[(4,5),(5,6),(4,5)]
var = [0,2,1]

for index, value in enumerate(env):
    plt.figure(figsize=dimen[index])
    plt.xlabel('Exposure Time (h)')
    plt.ylabel('Crystallinity Degree (%)')
    
    xmax = math.ceil(max(df_dsc[f"Tempo ({value})"].dropna()))
    
    ymin = math.floor(min(df_dsc[f"Cristalinidade ({value})"].dropna()))
    ymax = math.ceil(max(df_dsc[f"Cristalinidade ({value})"].dropna()))
    
    xint = [x for x in range(0, xmax + 50, 50)]
    yint = [y for y in range(ymin - delta_min[index], ymax + delta_max[index], inc[index])]
    
    plt.yticks(yint)
    plt.xticks(xint)
    
    plt.ylim(yint[0], yint[-1])
    plt.plot(df_dsc[f"Tempo ({value})"].dropna().astype(int),df_dsc[f"Cristalinidade ({value})"].dropna(),cor[index])
    plt.savefig(f'G:\\Meu Drive\\TCC\\resultados ensaios\\DSC\\Grafs_TCC_Images\\cristalinidade{value}.png', dpi = 500)
```
